# Q2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page1 = requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
page1

soup1 = BeautifulSoup(page1.content)
soup1

name = []
for i in soup1.find_all('h3',class_="lister-item-header"):
    name.append(i.text.replace('\n',' '))

rating = []

for i in soup1.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace('\n',''))
    
date=[]

for i in soup1.find_all('span',class_="lister-item-year text-muted unbold"):
    date.append(i.text)
    
import pandas as pd

df = pd.DataFrame({'Movie Name':name,'Rating':rating,'year of release':date})
df
```

# Q3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page2 = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page2

soup2 = BeautifulSoup(page2.content,'html.parser' )
soup2

name = []
for i in soup2.find_all('td',class_ ='titleColumn'):
        name.append(i.text.replace('\n',''))
        names_1=name[0:100]
        
rating=[]
for i in soup2.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',' '))
    rating_1=rating[0:100]
rating_1


date=[]
for i in soup2.find_all('span',class_="secondaryInfo"):
    date.append(i.text)
    date_1=date[0:100]
date_1


df1 = pd.DataFrame({'Movie Name':names_1,'Rating of movie':rating_1,'Year of release':date_1})
df1
```

# Q4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp

```python
page13=requests.get("https://meesho.com/bags-ladies/pl/p7vbp")
page13

soup13=BeautifulSoup(page13.content)
soup13

Product=[]
for i in soup13.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    Product.append(i.text.replace("'", "").lower())
Product

Price=[]
for i in soup13.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    Price.append(i.text.replace('₹',' '))
Price
Discount=[]
for i in soup13.find_all('span',class_ = 'Text__StyledText-sc-oo0kvp-0 lnonyH'):
    Discount.append(i.text.replace('%','').strip('').split(" ")[0])
Discount

df=pd.DataFrame({'Product Name':Product,'Price':Price,'Discount(%)':Discount})
df
```

# Q5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape

#  a)  Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page3 = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page3

soup3 = BeautifulSoup(page3.content)
soup3

name = []

for i in soup3.find_all('span',class_="u-hide-phablet"):
    name.append(i.text)
    team_name=name[0:10]
    
Points = []

for i in soup3.find_all('td',class_="rankings-block__banner--points"):
    Points.append(i.text)
for j in soup3.find_all('td',class_="table-body__cell u-center-text"):
    Points.append(j.text)
Points

del Points[1:21:2]
Points

points = Points[0:10]
points

Matches = []

for i in soup3.find_all('td',class_="rankings-block__banner--matches"):
    Matches.append(i.text)
for j in soup3.find_all('td',class_="table-body__cell u-center-text"):
    Matches.append(j.text)
    
Matches
del Matches[2:21:2]
Matches
matches = Matches[0:10]
matches

Rating=[]

for i in soup3.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    Rating.append(i.text.replace("\n",''))
for j in soup3.find_all('td',class_="table-body__cell u-text-right rating"):
    Rating.append(j.text)
    
rating=Rating[0:10]
rating

df2=pd.DataFrame({'Team Name':team_name,'Point':points,'Matches':matches,'Rating':rating})
df2

```

# Q7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
page9 = requests.get("https://coreyms.com/")
page9

soup9=BeautifulSoup(page9.content)
soup9

Heading=[]

for i in soup9.find_all('h2',class_="entry-title"):
    Heading.append(i.text)
    heading_1=Heading[0:9]
heading_1

Date=[]

for i in soup9.find_all('time',class_="entry-time"):
    Date.append(i.text)
    date_1=Date[0:9]
date_1

Content=[]

for i in soup9.find_all('div',class_="entry-content"):
    Content.append(i.text.replace('\n',''))
    content_1=Content[0:9]
content_1

youtube_url=[]
    
for i in soup9.find_all('iframe',class_ ="youtube-player"):
        youtube_url.append(i['src']) 
        
youtube_url

df8 = pd.DataFrame({'Heading':heading_1,'Date':date_1,'Content with url':content_1,'url':youtube_url})
df8
```

# Q9) Write a python program to scrape mentioned details from dineout.co.in 

# 1) Restaurant name

```python
page11=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page11

soup11=BeautifulSoup(page11.content)
soup11

Restaurant=[]
for i in soup11.find_all('div',class_="restnt-info cursor"):
    Restaurant.append(i.text)
Restaurant

Cuisine=[]
for i in soup11.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.replace('₹','').split('|')[1])
Cuisine

Location=[]
for i in soup11.find_all('div',class_ ='restnt-loc ellipsis'):
    Location.append(i.text)
Location

Rating=[]
for i in soup11.find_all('div',class_='restnt-rating rating-4'):
    Rating.append(i.text)
Rating

Image_url=[]
for i in soup11.find_all('img',class_='no-img'):
    Image_url.append(i['data-src'])
Image_url

df = pd.DataFrame({'Restaurant name':Restaurant,'Cuisine':Cuisine,'Location':Location,'Rating':Rating,'image_url':Image_url})
df
    
```

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
page12 = requests.get("https://www.bewakoof.com/women-tshirts?ga_q=tshirts") 
page12

soup12=BeautifulSoup(page12.content)
soup12

product= []
for i in soup12.find_all('div',class_="productCardDetail"):
    product.append(i.text.split('₹')[0])
product 

price = []
tribe_member = []
for i in soup12.find_all('div',class_='productPriceBox clearfix'):
    price.append(i.text.split('₹')[1].replace('FEW LEFT',""))
    tribe_member.append(i.text.split('₹')[-1].replace('For TriBe Members',''))
price
tribe_member

image_url = []
for i in soup12.find_all('img',class_='productImgTag'):
        image_url.append(i['src'])
image_url

df = pd.DataFrame({'Product Name':product,'Price':price,'Tribe member price':tribe_member,'image_url':image_url})
df
```
